```python
# Read in the file
zz=list()
with open('input.txt', 'r') as _f:
    for l in _f:
        z = list(l.strip())
        zz.append(z)
print(zz)
```

```python
# Algorithm
cnt=0
w= len(zz[0])
l= len(zz)
for i in range(0,l):
    for j in range(0,w):
        if (i==0) | (j==0) | (i==w-1) | (j==l-1):
                cnt+=1
                continue
        else:
            ref = zz[i][j]
            # Look to right
            visible =True
            for x in range(j+1,w):
                visible = visible and (zz[i][x]<ref)
            if visible:
                cnt+=1
                continue
            # Look to the left
            visible=True
            for x in range(0,j):
                visible = visible and (zz[i][x]<ref)
            if visible:
                cnt+=1
                continue
            # Look down
            visible=True
            for y in range(i+1,l):
                visible = visible and (zz[y][j]<ref)
            if visible:
                cnt+=1
                continue
            visible=True
            for y in range(0,i):
                visible = visible and (zz[y][j]<ref)
            if visible:
                cnt+=1
                continue

print(cnt)
```

```python
# Algorithm
vd=0
vdu=0
vdd=0
vdl=0
vdr=0
w= len(zz[0])
l= len(zz)
vdm = 0
for i in range(0,l):
    for j in range(0,w):
        if (i==0) | (j==0) | (i==w-1) | (j==l-1):
                vdu=0
                continue
        else:
            ref = zz[i][j]
            # Look to right
            vdr=0
            for x in range(j+1,w):
                vdr+=1
                if (zz[i][x]>=ref):
                    break

            # Look to the left
            vdl=0
            for x in range(j-1,-1,-1):
                vdl+=1
                if (zz[i][x]>=ref):
                    break
            # Look down
            vdd = 0
            for y in range(i+1,l):
                vdd+=1
                if (zz[y][j]>=ref):
                    break
            vdu = 0
            for y in range(i-1,-1,-1):
                vdu+=1
                if (zz[y][j]>=ref):
                    break
        vd = vdl*vdr*vdu*vdd
        #print(i,j,"---",vdr,vdl,vdd,vdu)
        if vd>vdm:
            vdm=vd
            zzz=(i,j)

                    
print(vdm)
print(zzz)
```
